```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("cloudburst.csv")
df.head()
```

```python
df.dtypes
```

```python
df1 = pd.get_dummies(data = df,columns=['WindGustDir','Date','Location','WindDir3pm','WindDir9am'])
df1
```

```python
def print_uniq_val(df):
    for column in df:
        print(f'{column} : {df[column].unique()}')
```

```python
print_uniq_val(df1)
```

```python
df1.replace('Yes',1,inplace=True)
df1.replace('No',0,inplace=True)
df1.head()
```

```python
print_uniq_val(df1)
```

```python
cols_to_scale = ['MinTemp','MaxTemp','Rainfall','Evaporation','Sunshine','WindGustSpeed','WindSpeed9am','WindSpeed3pm','Humidity9am','Humidity3pm','Pressure9am','Cloud9am','Cloud3pm','Temp9am','Temp3pm']
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df1[cols_to_scale] = scaler.fit_transform(df1[cols_to_scale])
df1.head()
```

```python
print_uniq_val(df1)
```

```python
x = df1.drop('CloudBurst Today',axis='columns')
y = df1['CloudBurst Today']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2022)
```

```python
print_uniq_val(df1)
```

```python
x_train = np.asarray(x_train).astype(np.float32)
y_train = np.asarray(y_train).astype(np.float32)
```

```python
import tensorflow as tf
from tensorflow import keras


model = keras.Sequential([
    keras.layers.Dense(3000, input_shape=(3550,), activation='relu'),
    keras.layers.Dense(2700, activation='relu'),
    keras.layers.Dense(2400, activation='relu'),
    keras.layers.Dense(2100, activation='relu'),
    keras.layers.Dense(1800, activation='relu'),
    keras.layers.Dense(1500, activation='relu'),
    keras.layers.Dense(1200, activation='relu'),
    keras.layers.Dense(900, activation='relu'),
    keras.layers.Dense(600, activation='relu'),
    keras.layers.Dense(300, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.fit(x_train, y_train, epochs=1)
```

```python
model.evaluate(x_test, y_test)
```
